**Sobre este curso:** 
Este curso no pretende enseñar todo sobre el manejo de Python, pero busca una comprensión profunda 
de los conceptos detrás de la herramienta y su uso en la solución de problemas reales. Al final de 
este curso, el estudiante pueda continuar su formación de forma autónoma.

**Descripción:** 
La ciencia de datos es un campo interdisciplinario que involucra métodos científicos, 
procesos y sistemas para extraer conocimiento o un mejor entendimiento de datos en sus diferentes formas, 
ya sea estructurados o no estructurados,​ lo cual es una continuación de algunos campos de análisis de datos 
como la estadística, la minería de datos, el aprendizaje automático, y la analítica predictiva. 

**Objetivos:** 
Al finalizar el modulo el estudiante habrá conocido los principales aspectos del lenguaje de programación Python, sus características, 
modos de instalación, funcionamiento básico, soporte comunitario y los recursos mas destacados disponibles 
en la Web para tener en cuenta al momento de enfrentarse a problemas de ciencia de datos. 

**Conceptos previos necesarios:** 
Este curso es autocontenido, pero la experiencia previa en programación es deseable.

**Calendario y temas cubiertos:** 
El curso está dividido en cinco (9) bloques temáticos virtuales. Cada bloque está compuesto por una sesión virtual  
de 90 minutos (1.5 horas) y su correspondiente evaluación. Adicionalmente, 
durante las sesiones se pondrán a disposición de los estudiantes material adicional con el fin de que puedan profundizar sobre aspectos 
puntuales de los temas desarrollados. Los temas cubiertos están disponibles en [cursos](https://dnajarm.github.io/courses/introduccion_a_la_ciencia_de_datos/index.html)

**Metodología:** 
En este modulo el trabajo individual por parte del estudiante es fundamental para el logro de los objetivos. 
El curso se desarrollará utilizando los siguientes instrumentos pedagógicos:

- Conferencias magistrales virtuales por parte del docente.
- Demostraciones funcionales de Python, 
- Talleres individuales y/o grupales.
- Laboratorios para solución de problemas de programación por parte del estudiante.
- Estudio individual por parte del estudiante.

Las sesiones están diseñadas para que el estudiante desarrolla conceptos fundamentales sólidos sobre los temas tratados. 
El estudiante debe esforzarse por reforzar dichos conocimientos a partir de la lectura del material sugerido y la elaboración de ejercicios y laboratorios de programación.
